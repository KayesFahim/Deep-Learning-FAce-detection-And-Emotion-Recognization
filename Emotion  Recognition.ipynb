# Import All image and store in a array

```python
Training_data = []

for fle in files:
  idx=Classes.index(fle)
  label=idx
  
  total=Datadir+'/'+fle
  files_exp= os.listdir(total)
  for fle_2 in files_exp:
    file_main=total+'/'+fle_2
    print(file_main+"   "+str(label))
    image= cv2.imread(file_main)
    img_array = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    new_array= cv2.resize(image,(img_size,img_size))
    Training_data.append([new_array,label])
```

```python
print(len(Training_data))
```

```python
temp = np.array(Training_data, dtype=object)
```

```python
temp.shape
```

```python
import random

random.shuffle(Training_data)
```

```python
x = [] #data/features
y = [] #labels

for features,label in Training_data:
    x.append(features)
    y.append(label)
X = np.array(x).reshape(-1,img_size,img_size,3)
```

```python
X.shape
```

```python
#Normalize data

X= X/255.0;
```

```python
type(y)
```

```python
Y = np.array(y)
```

```python
Y.shape
```

# Deep Learning Model

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
```

```python
model = tf.keras.applications.VGG19() #Pre-Trained Model
```

```python
model.summary()
```

```python
base_input = model.layers[0].input
```

```python
base_output = model.layers[-2].output
```

```python
base_output
```

```python
final_output = layers.Dense(128)(base_output)
final_ouput = layers.Activation('relu')(final_output)
final_output = layers.Dense(64)(final_ouput) 
final_ouput = layers.Activation('relu')(final_output) 
final_output = layers.Dense(7,activation='softmax')(final_ouput) 
```

```python
final_output
```

```python
new_model = keras.Model(inputs = base_input, outputs=final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics= ["accuracy"])
```

```python
new_model.fit(X,Y, epochs=100)
```

```python
new_model.save("Final_model_epoch100_vgg.h5")
```

```python
new_model = tf.keras.models.load_model('Final_model_epoch100.h5')
```

```python
new_model.evaluate
```

# Detect Image

```python
frame = cv2.imread("happyness.jpg")
```

```python
# load image from file
frame.shape
```

```python
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
FaceCascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
grey = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
grey.shape
```

```python
faces = FaceCascade.detectMultiScale(grey,1.1,4)
```

```python
for x,y,w,h in faces:
    roi_grey = grey[y:y+h, x:x+w]
    roi_color = frame[y:y+h, x:x+w]
    cv2.rectangle(frame, (x,y), (x+w, y+h), (255,0,0),2)
    facess = FaceCascade.detectMultiScale(roi_grey)
    if len(facess) == 0:
        print("faces Not detected")
    else:
        for (ex, ey, ew, eh) in facess:
            face_roi = roi_color[ey:ey+eh, ex:ex + ew]
            
```

```python
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(face_roi, cv2.COLOR_BGR2RGB))
```

```python
final_image = cv2.resize(face_roi, (224,224))
final_image = np.expand_dims(final_image, axis =0)
final_image = final_image/255.0
```

```python
Predictions = new_model.predict(final_image)
```

```python
Predictions[0]
```

```python
Result = np.argmax(Predictions)
```

```python
#Classes = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
if Result == 0:
    print("angry")
elif  Result == 1:
    print("disgust")
elif  Result == 2:
    print("fear")
elif  Result == 3:
    print("happy")
elif  Result == 4:
    print("neutral")
elif  Result == 5:
    print("sad")
elif  Result == 6:
    print("surprise")
else:
    print("Invalid")
```
